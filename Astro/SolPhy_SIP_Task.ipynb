```python
import astropy.io.fits as fits
import cv2
import numpy as np
import matplotlib.pyplot as plt


with fits.open('HA_19490104T081200_Q1L0a.fits') as hdul:
    image_data = hdul[0].data
    

normalized_image = ((image_data - image_data.min()) / (image_data.max() - image_data.min()) * 255)
jpeg_path = 'output_image.jpg' 
cv2.imwrite(jpeg_path, normalized_image)

sun = cv2.imread('output_image.jpg')
helios = cv2.imread('output_image.jpg')
gray_img = cv2.cvtColor(sun, cv2.COLOR_BGR2GRAY)
img = cv2.medianBlur(gray_img, 5)
cimg = cv2.cvtColor(img,cv2.COLOR_GRAY2BGR)

circles = cv2.HoughCircles(img,cv2.HOUGH_GRADIENT,1,120,param1=100,param2=30,minRadius=1,maxRadius=2000)
circles = np.uint16(np.around(circles))

for i in circles[0,:]:
   # draw the outer circle
   cv2.circle(sun,(i[0],i[1]),i[2],(0,255,0),5)
   # draw the center of the circle
   cv2.circle(sun,(i[0],i[1]),5,(255,0,0),5)
   print("\n\n\nRadius of the sun:",i[2])
   print("Coordinates of the center:({},{})  (In pixel-coordinates)\n\n\n".format(i[0],i[1]))


ret, bw_img = cv2.threshold(helios, 100, 255, cv2.THRESH_BINARY)     

cv2.imwrite("Limb_detected_sun.jpg", sun)
cv2.imwrite("Binary_sun.jpg", bw_img)

fig, axs = plt.subplots(1,2, figsize=(10,5))
axs[0].imshow(sun)
axs[0].set_title("Limb detected sun")
axs[1].imshow(bw_img)
axs[1].set_title("Binary image")

plt.tight_layout()
plt.show()
```

```python
import astropy.io.fits as fits
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Function to apply watershed segmentation for sunspot detection
def detect_sunspots(image):
    # Convert the image to grayscale
    gray_img = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    # Apply histogram equalization to enhance contrast
    equ_img = cv2.equalizeHist(gray_img)

    # Apply thresholding to create a binary mask of potential sunspots
    _, thresh = cv2.threshold(equ_img, 0, 255, cv2.THRESH_BINARY_INV + cv2.THRESH_OTSU)

    # Perform morphological operations to clean up the mask
    kernel = np.ones((3,3), np.uint8)
    opening = cv2.morphologyEx(thresh, cv2.MORPH_OPEN, kernel, iterations=2)

    # Use distance transform and watershed segmentation to separate sunspots
    dist_transform = cv2.distanceTransform(opening, cv2.DIST_L2, 5)
    _, sure_fg = cv2.threshold(dist_transform, 0.7 * dist_transform.max(), 255, 0)
    sure_fg = np.uint8(sure_fg)

    return sure_fg

# Load FITS image and preprocess
with fits.open('UDAI.FDGB.01032023.043724.015.fits') as hdul:
    image_data = hdul[0].data

normalized_image = ((image_data - image_data.min()) / (image_data.max() - image_data.min()) * 255).astype('uint8')
jpeg_path = 'output_image.jpg' 
cv2.imwrite(jpeg_path, normalized_image)

sun = cv2.imread('output_image.jpg')

# Detect sunspots using watershed segmentation
sunspots_binary = detect_sunspots(sun)

# Find and draw circles around the sun
gray_img = cv2.cvtColor(sun, cv2.COLOR_BGR2GRAY)
img = cv2.medianBlur(gray_img, 5)
circles = cv2.HoughCircles(img, cv2.HOUGH_GRADIENT, 1, 120, param1=100, param2=30, minRadius=700, maxRadius=1250)
if circles is not None:
    circles = np.uint16(np.around(circles))
    for i in circles[0, :]:
        # draw the outer circle
        cv2.circle(sun, (i[0], i[1]), i[2], (0, 255, 0), 5)
        # draw the center of the circle
        cv2.circle(sun, (i[0], i[1]), 5, (255, 0, 0), 5)
        print("Radius of the sun:", i[2])
        print("Coordinates of the center: ({},{})  (In pixel-coordinates)".format(i[0], i[1]))

# Display and save the results
cv2.imwrite("Sunspots_binary.jpg", sunspots_binary)
cv2.imwrite("Sun_with_circles.jpg", sun)

fig, axs = plt.subplots(1, 2, figsize=(10, 5))
axs[0].imshow(cv2.cvtColor(sun, cv2.COLOR_BGR2RGB))
axs[0].set_title("Sun with Circles")
axs[0].axis('off')
axs[1].imshow(sunspots_binary, cmap='gray')
axs[1].set_title("Binary Sunspots Image")
axs[1].axis('off')

plt.tight_layout()
plt.show()
```
